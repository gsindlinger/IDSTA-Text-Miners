## Extracting text using PyPDF2

```python
import PyPDF2
```

```python
reader = PyPDF2.PdfReader("liste1.pdf")
pages = reader.pages
output = ""
for page in pages:
    output += "\n" + page.extract_text()

##saving as txt file
with open('liste1_pypdf2.txt', 'w') as f:
    f.write(output)
```

## Extracting text using OCR
##### we will first convert the PDF to images that can be analysed using the py-tesseract

```python
# import module
from pdf2image import convert_from_path
 
# Store Pdf with convert_from_path function
images = convert_from_path('liste1.pdf')
 
for i in range(len(images)):
      # Save pages as images in the pdf
    images[i].save('page'+ str(i) +'.jpg', 'JPEG')
```

#### we will now pass those images to the pytesseract library

```python
import cv2
import pytesseract
from pytesseract import Output

imgs = ["page0.jpg", "page1.jpg"]

output = ''
for img in imgs:
    print(img)
    read_img = cv2.imread(img)
    d = pytesseract.image_to_string(read_img, lang='deu')
    print(d)
    output += '\n' + d

img = cv2.imread("page1.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("page 2 : " , pytesseract.image_to_string(gray))
    
##saving as txt file
with open('liste1_ocr.txt', 'w') as f:
    f.write(output)
```

## Comparison of the two output files using SequenceMatcher

```python
from difflib import SequenceMatcher
with open('liste1.ocr.txt', 'r') as f:
    liste1_ocr = f.read()
with open('liste1_pypdf2.txt', 'r') as f:
    liste1_pypdf2 = f.read() 
SequenceMatcher(a = liste1_pypdf2, b = liste1_ocr).ratio()
```

## Extracting IBANs out of pypdf2 text file

```python
with open("liste1_pypdf2.txt", mode='r') as f:
    data = f.read()
print(data)
```

```python
import regex as re
pattern = r'(([A-Z]{2}[0-9]{2}) ([A-Z[0-9]+\s)+)'
ibans = re.findall(pattern, data)
print(ibans)
ibans_text = ""
for iban in ibans:
    ibans_text += iban[0]
with open("ibans_list.txt", "w", encoding = "utf8") as f:
    f.write(ibans_text)
```
