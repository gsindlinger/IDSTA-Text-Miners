## Extracting text using PyPDF2

```python
import PyPDF2
```

```python
reader = PyPDF2.PdfReader("bahnstadt.pdf")
pages = reader.pages
output = ""
for page in pages:
    output += "\n" + page.extract_text()

##saving as txt file
with open('bahnstadt_pypdf2.txt', 'w', encoding = 'utf8') as f:
    f.write(output)
```

## Extracting text using OCR
##### we will first convert the PDF to images that can be analysed using the py-tesseract

```python
## PYTHON IS WEIRD
import os
current_path = os.getcwd()
current_path
os.chdir(current_path + "\\bahnstadt")
new_path = os.getcwd()
new_path
```

```python
# import module
from pdf2image import convert_from_path
 
# Store Pdf with convert_from_path function
images = convert_from_path('bahnstadt.pdf')
read_images = []
for i in range(len(images)):
      # Save pages as images in the pdf
    images[i].save('page'+ str(i) +'.jpg', 'JPEG')
    read_images.append('page' + str(i) + '.jpg') ## setting up a list of the file names to read later
```

## we will now pass those images to the pytesseract library

```python
import cv2
import pytesseract
from pytesseract import Output

output = ''
for image in read_images:
    print(type(image))
    read_img = cv2.imread(image)
    gray = cv2.cvtColor(read_img, cv2.COLOR_BGR2GRAY)
    d = pytesseract.image_to_string(read_img, lang='deu')
    print(d)
    output += '\n' + d
    
##saving as txt file
with open('bahnstadt_ocr.txt', 'w', encoding = 'utf8') as f:
    f.write(output)
```

## Comparison of the two output files using SequenceMatcher

```python
from difflib import SequenceMatcher
with open('bahnstadt_ocr.txt', 'r' , encoding = 'utf8') as f:
    bahnstadt_ocr = f.read()
with open('bahnstadt_pypdf2.txt', 'r', encoding = 'utf8') as f:
    bahnstadt_pypdf2 = f.read()
print(SequenceMatcher(a = bahnstadt_pypdf2,b = bahnstadt_ocr).ratio())
```

## Opening the pypdf2 extracted text and getting information

```python
import regex as re
with open("bahnstadt_pypdf2.txt", mode='r', encoding = 'utf8') as f:
    data = f.read()
```

```python
## find all emails
pattern = r'[a-zA-Z][a-zA-Z0-9._+-]+@[A-Za-z\-\.]+.[A-Za-z]+?\s'
emails = re.findall(pattern, data)
print(emails)
emails_text = "" 
for email in emails:
    emails_text += email + "\n"
with open("bahnstadt_pypdf2_emails.txt", "w", encoding = "utf8") as f:
    f.write(emails_text)
```

```python
## find all URLS
pattern = r'[www]+\.+[a-zA-Z+\-]+\.+[a-zA-Z]{1,3}'
urls = re.findall(pattern, data)
print(urls)
urls_text = ""
for url in urls:
    urls_text += url + "\n"
with open("bahnstadt_pypdf2_urls.txt", "w", encoding = "utf8") as f:
    f.write(urls_text)
```

```python
## find all Phone Numbers
## https://regex101.com/r/F4A99G/1 - for some reason works here but not in python
## fixed now works 
## https://regex101.com/r/F4A99G/2
pattern = r'0[0-9/.\s]{2,7}[^a-zA-Z\,]+'
numbers = re.findall(pattern, data)
numbers_text = ""
print(numbers)
for number in numbers:
    numbers_text += number + "\n"
print(numbers_text)
with open("bahnstadt_pypdf2_numbers.txt", "w", encoding = "utf8") as f:
    f.write(numbers_text)
```
