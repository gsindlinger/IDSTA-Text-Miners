```python
# importing required modules 
from PyPDF2 import PdfReader
import regex as re
```

## Extract using pypdf2

```python
reader = PdfReader('bundeswehr.pdf')
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = ""
for page in reader.pages:
    text += page.extract_text()
    
with open('bundeswehr_pypdf2.txt', 'w', encoding = 'utf8') as f:
    f.write(text)
```

## Optimizing output for better reading later on

```python
## remove all \n
text = text.replace("\n", "")
text
```

```python
## adding a blank space before http://
## remove all \n
text = text.replace("http://", " http://")
text
```

```python
## Website addresses and +49 are mostly connected to each other, along other instances of +49 being connected to another part of the string
## remove all \n
text = text.replace("+49", " +49")
```

```python
with open("bundeswehr_pypdf2_optimized.txt", "w", encoding = 'utf8') as f:
    f.write(text)
```

## Extracting using pytesseract (OCR)

#### converting all pages to jpg files

```python
# import module
from pdf2image import convert_from_path
 
# Store Pdf with convert_from_path function
images = convert_from_path('bundeswehr.pdf')
read_images = []
for i in range(len(images)):
      # Save pages as images in the pdf
    images[i].save('page'+ str(i) +'.jpg', 'JPEG')
    read_images.append('page' + str(i) + '.jpg') ## setting up a list of the file names to read later
```

```python
import cv2
import pytesseract
from pytesseract import Output

output = ''
for image in read_images:
    print(type(image))
    read_img = cv2.imread(image)
    gray = cv2.cvtColor(read_img, cv2.COLOR_BGR2GRAY)
    d = pytesseract.image_to_string(read_img, lang='deu')
    print(d)
    output += '\n' + d
    
##saving as txt file
with open('bundeswehr_ocr.txt', 'w', encoding = 'utf8') as f:
    f.write(output)
```

```python
from difflib import SequenceMatcher

with open('bundeswehr_ocr.txt', 'r', encoding = 'utf8') as f:
    bundeswehr_ocr = f.read()
with open('bundeswehr_pypdf2.txt', 'r', encoding = 'utf8') as f:
    bundeswehr_pypdf2 = f.read()
    
SequenceMatcher(a = bundeswehr_pypdf2, b = bundeswehr_ocr).ratio()
```

## Extracting URLS, Email adresses and phone numbers from the file

```python
with open("bundeswehr_pypdf2_optimized.txt", mode='r') as f:
    data = f.read()
data
```

```python
## find all emails
pattern = r'[a-zA-Z][a-zA-Z0-9._+-]+@[A-Za-z\-\.]+.[A-Za-z]+?\s'
emails = re.findall(pattern, data)
emails_text = "" 
for email in emails:
    emails_text += email + "\n"
with open("bundeswehr_pypdf2_emails.txt", "w", encoding = "utf8") as f:
    f.write(emails_text)
```

```python
## find all URLS
pattern = r'(http|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])'
urls = re.findall(pattern, data)
urls_text = ""
for url in urls:
    urls_text += url[1] + url[2] + "\n"
with open("bundeswehr_pypdf2_urls.txt", "w", encoding = "utf8") as f:
    f.write(urls_text)
```

```python
## find all Phone Numbers
## https://regex101.com/r/F4A99G/1 - for some reason works here but not in python
## fixed now works 
## https://regex101.com/r/F4A99G/2
pattern = r'(\+49+(([\s\-])+[0-9]+)+)'
numbers = re.findall(pattern, data)
numbers_text = ""
for number in numbers:
    numbers_text += number[0] + "\n"
print(numbers_text)
with open("bundeswehr_pypdf2_numbers.txt", "w", encoding = "utf8") as f:
    f.write(numbers_text)
```
