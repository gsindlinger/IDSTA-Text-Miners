```python
import pandas as pd
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
# use tfidf by removing tokens that don't appear in at least 50 documents
vect = TfidfVectorizer(min_df=100)
# Fit and transform
X = vect.fit_transform(clean_list)
```

```python
# Create an NMF instance: model
# the 10 components will be the topics
model = NMF(n_components=10, random_state=5)
# Fit the model to TF-IDF
model.fit(X)
# Transform the TF-IDF: nmf_features
nmf_features = model.transform(X)
```

```python
# Create a DataFrame: components_df
components_df = pd.DataFrame(model.components_, columns=vect.get_feature_names())
components_df
```

```python
import gensim

model = gensim.models.Word2Vec(ls, min_count=50, workers = 10, window = 5, vector_size = 200)
```

```python
import smart_open

def read_corpus(fname, tokens_only=False):
    with smart_open.open(fname, encoding="iso-8859-1") as f:
        for i, line in enumerate(f):
            tokens = gensim.utils.simple_preprocess(line)
            if tokens_only:
                yield tokens
            else:
                # For training data, add tags
                yield gensim.models.doc2vec.TaggedDocument(tokens, [i])

train_corpus = list(read_corpus("lines_list.txt"))
test_corpus = list(read_corpus("lines_list.txt", tokens_only=True))

doc_model = gensim.models.doc2vec.Doc2Vec(vector_size=200, min_count=50, epochs=40)
```

```python
doc_model.build_vocab(train_corpus)
doc_model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
ranks = []
second_ranks = []
for doc_id in range(len(train_corpus)):
    inferred_vector = doc_model.infer_vector(train_corpus[doc_id].words)
    sims = doc_model.dv.most_similar([inferred_vector], topn=len(doc_model.dv))
    rank = [docid for docid, sim in sims].index(doc_id)
    ranks.append(rank)
    second_ranks.append(sims[1])
```
